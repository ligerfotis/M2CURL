```python
# open saved_models/object_push-v0/20240117-194243/mvitac_sac/s1_visuotactile/trained_models/evaluations.npz file
# and plot the results
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
# load the data
# file_path = 'saved_models/object_push-v0/20240117-194243/mvitac_sac/s1_visuotactile/trained_models/evaluations.npz'
file_path = 'saved_models/object_push-v0/20240118-000555/mvitac_sac/s1_visuotactile/trained_models/evaluations.npz'
# Load the data from the NPZ file
data = np.load(file_path)
```

```python
data.keys()
```

```python
# Extracting data from the file
timesteps = data['timesteps']
results = data['results']
ep_lengths = data['ep_lengths']
```

```python
# Display the shapes of loaded arrays for verification
print("Timesteps Shape:", timesteps.shape)
print("Results Shape:", results.shape)
print("Episode Lengths Shape:", ep_lengths.shape)
```

```python
print("Timesteps:", timesteps[0])
print("Results:", results[0])
print("Episode Lengths:", ep_lengths[0])
```

```python
# Assuming the data is already loaded into 'timesteps', 'results', and 'ep_lengths'
# Calculate the mean and standard deviation of rewards
# rewards = results[0]
mean_rewards = results.mean(axis=1)
std_rewards = results.std(axis=1)
# x_axis = np.arange(0, timesteps[0], 100000)
# Plotting the rewards
plt.figure(figsize=(10, 6))
plt.plot(timesteps, mean_rewards, label='Mean Reward')
plt.fill_between(timesteps, mean_rewards - std_rewards, mean_rewards + std_rewards, alpha=0.2, label='Standard Deviation')
plt.xlabel('Timesteps')
plt.ylabel('Rewards')
plt.title('Mean Rewards Over Timesteps with Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```
